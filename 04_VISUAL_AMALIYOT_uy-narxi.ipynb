```python
df = pd.read_excel('/kaggle/input/tashkent-real-estate-2019/uybor.xlsx')
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.figure(figsize=(12,5))
sns.displot(data=df, x='price', height=4, aspect=2)
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15,10))
# Umumiy chizma nomini beramiz:
fig.suptitle("uylarni narxlash")
sns.histplot(ax=ax[0, 0], data=df, x='price')
sns.histplot(ax=ax[0, 1], data=df, x='size')
sns.scatterplot(ax=ax[1, 0], data=df, x=df['rooms'], y=df['price'], hue="level", s=50)
sns.scatterplot(ax=ax[1, 1], data=df, x=df['rooms'], y=df['size'], hue="level", s=50)
# Har bir grafik uchun nom:
ax[0,0].set_title("narxi boyicha")
ax[0,1].set_title("maydoni boyicha")
ax[1,0].set_title("narxi va honalar soni bogliqligi")
ax[1,1].set_title("maydoni va hona soni bilan bogliqligi")
plt.show()
```
